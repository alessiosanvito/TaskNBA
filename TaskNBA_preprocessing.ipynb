# Modifiche

```python
gamesNA = games[games['PTS_away'].isna()]
```

```python
games = games.drop(index=gamesNA.index)
```

```python
games.to_csv('games.csv', index=False)
```

```python
games_detailsNA = games_details[games_details['MIN'].isna()]
```

```python
games_details = games_details.drop(index=games_detailsNA.index)
```

```python
games_details00 = games_details[games_details['MIN'] == '0:00']
```

```python
games_details = games_details.drop(index=games_details00.index)
```

```python
games_details.to_csv('games_details.csv', index=False)
```

```python
# Effettua la join sulla colonna 'GAME_ID'
merged_df = pd.merge(games_details, games[['GAME_ID', 'SEASON']], on='GAME_ID', how='left')

merged_df
```

```python
# Dividere la colonna 'MIN' in due colonne 'MINUTES' e 'SECONDS'
merged_df[['MINUTES', 'SECONDS']] = merged_df['MIN'].str.split(':', expand=True)
```

```python
# Conversione delle colonne in tipo numerico, se necessario
merged_df['MINUTES'] = pd.to_numeric(merged_df['MINUTES'])
merged_df['SECONDS'] = pd.to_numeric(merged_df['SECONDS']).fillna(0)
```

```python
# Calcolare la durata totale delle partite in minuti
merged_df['TOTAL_MINUTES'] = merged_df['MINUTES'] + merged_df['SECONDS'] / 60
```

```python
merged_df
```

```python
merged_df.to_csv('games_details_seas.csv', index=False)
```

```python
# Modifiche a teams.csv

# Mappa dei valori da assegnare in base alla colonna 'CITY'
capacity_mapping = {
    'Brooklyn': 17732,
    'Orlando': 20000,
    'Philadelphia': 21000,
    'Phoenix': 18422,
    'New Orleans': 17791
}

# Sostituisci i valori NaN in base alla mappa
teams['ARENACAPACITY'] = teams.apply(lambda row: capacity_mapping[row['CITY']] if pd.isna(row['ARENACAPACITY']) else row['ARENACAPACITY'], axis=1)
```

```python
teams.to_csv('teams.csv', index=False)
```
